```
import torch
```

```
#AUTOGRAD
x=torch.ones(5, 3, requires_grad=True)
print(x)
print(x.grad_fn)
y=x+2
print(y)
print(y.grad_fn)
```

```
z=y*y*5
print(z)
out=z.mean()  #calculates the avg of all elements
print(out)
```

```
# changing required grad status
a=torch.rand(5, 3)
a=((a*8)/(a-5))
print(a)
print(a.requires_grad)
a.requires_grad=True
print(a.requires_grad)
b=(a*a).sum()
print(b)
print(b.grad_fn)
```

```
#getting grads
out.backward()
print(x.grad)
```

```
#VECTOR-JACOBIAN
x=torch.randn(3, requires_grad=True)
print('x:', x)
y=x*2
while y.data.norm() < 1000:                        #return L2 norm of the array
  y=y*5
  print(y)

 

```

```
#calculating vector jacobian product

v=torch.tensor([0.02, 5.67, 8.00], dtype=torch.float)

y.backward(v, retain_graph=True)

print(x.grad)
```

```
#stopping autograd.

print(x.requires_grad)
print((x**2).requires_grad)

with torch.no_grad():
  print((x**2).requires_grad)
```

```
print(x.requires_grad)
y=x.detach()
print(y.requires_grad)
print(x.eq(y).all())
```
