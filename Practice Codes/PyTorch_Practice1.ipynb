```
import torch
from __future__ import print_function
import numpy as np
```

```
empty_arr=torch.empty(5, 3)         #create empty array
print(empty_arr)
```

```
rand_arr=torch.rand(5, 3)    #create random array
print(rand_arr)
```

```
zeros_arr=torch.zeros(5, 3, dtype=torch.long)   #create zero array
print(zeros_arr)
```

```
self_arr=torch.tensor([5, 556.86])    #create self defined array
print(self_arr)
```

```
x=self_arr.new_ones(5, 3, dtype=torch.int)
print(x)
print('size of x: ', x.size())

y=torch.randn_like(x, dtype=torch.float)
print(y)
print('size of y: ', y.size())
```

```
a=torch.ones(5, 3, dtype=torch.float)
print(a)
b=torch.rand(5, 3)
print(b)
```

```
#operations on arrays
#different ways for addition
c=a+b
print(c)
d=torch.add(a, b)
print(d)
e=torch.empty(5, 3)
torch.add(a, b, out=e)
print(e)
b.add_(a)
print(b)
```

```
#accessing array like in numpy
print(b[:, 2])
```

```
#resizing arrays
f=torch.rand(4, 5)
g=f.view(20)
h=f.view(10, 2)
print(f.size(), g.size(), h.size())
```

```
#to get python no. from tensor value
i=torch.rand(1)
print(i)
print(i.item())
```

```
# NumPy Bridge : Converting from tensor to NumPy and vice versa

torch_arr_1=torch.ones(5, 3)
numpy_arr_1=torch_arr_1.numpy()
print(torch_arr_1)
print(numpy_arr_1)
```

```
#IMOPRTANT THING HERE
one_arr=torch.ones(5, 3)
torch_arr_1.add_(one_arr)
#here changing torch arr also changes the numpy arr also ***************************
print(torch_arr_1)
print(numpy_arr_1)
```

```
#from numpy to torch array
np_arr=np.ones((5, 3))
torch_arr=torch.from_numpy(np_arr)
np.add(np_arr, np.ones((5, 3)), out=np_arr)
print(np_arr)
print(torch_arr)
```

```
#CUDA Tensors
# let us run this cell only if CUDA is available
# We will use ``torch.device`` objects to move tensors in and out of GPU
if torch.cuda.is_available():
    device = torch.device("cuda")          # a CUDA device object
    y = torch.ones_like(x, device=device)  # directly create a tensor on GPU
    x = x.to(device)                       # or just use strings ``.to("cuda")``
    z = x + y
    print(z)
    print(z.to("cpu", torch.double))       # ``.to`` can also change dtype together!
```
